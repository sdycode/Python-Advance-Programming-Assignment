2. 2.  Make a function that encrypts a given input with these steps:

Input: "apple"

Step 1: Reverse the input: "elppa"

Step 2: Replace all vowels using the following chart:

a => 0
e => 1
i => 2
o => 2
u => 3

 "1lpp0"

Step 3: Add "aca" to the end of the word: "1lpp0aca"

Output: "1lpp0aca"

Examples

encrypt("banana") ➞ "0n0n0baca"

encrypt("karaca") ➞ "0c0r0kaca"

encrypt("burak") ➞ "k0r3baca"

encrypt("alpaca") ➞ "0c0pl0aca"


```python

d = {'a':'0','e':'1','i':'2','o':'2','u':'3'}
def encrypt(s):
  s = s[::-1]
  for i in range(len(s)):
    if s[i] in d.keys():      
      s =s.replace(s[i], str(d.get(s[i])))
    
  s =s+'aca'
  print(s)

encrypt("banana")
encrypt("karaca")
encrypt("burak") 
encrypt("alpaca")  
```

3. 3. Given the month and year as numbers, return whether that month contains a Friday 13th.(i.e You can check Python's datetime module)

Examples

has_friday_13(3, 2020) ➞ True

has_friday_13(10, 2017) ➞ True

has_friday_13(1, 1985) ➞ False


```python
import datetime

def has_friday_13(m, y):
  return True if datetime.date(y,m,13).weekday() == 4 else False
 
print(has_friday_13(3, 2020))
print(has_friday_13(10, 2017))
print(has_friday_13(1, 1985) )
```

5. 5. Given a list of words in the singular form, return a set of those words in the plural form if they appear more than once in the list.

Examples

pluralize(["cow", "pig", "cow", "cow"]) ➞ { "cows", "pig" }

pluralize(["table", "table", "table"]) ➞ { "tables" }



```python

def pluralize(l):
  
  finalList = list(set(l))
  d ={}
  for i in finalList:
    d[i]=0
  for i in l:      
    d[i] += 1
  for i,e in enumerate(finalList):
    if(d[e]>1):
      finalList[i]= e+'s'

  print(finalList)

pluralize(["cow", "pig", "cow", "cow"])

pluralize(["table", "table", "table"]) 
  
```
